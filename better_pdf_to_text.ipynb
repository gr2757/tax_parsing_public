```python
import os
import subprocess

# Path to the pdftotext executable
pdftotext_path = '/opt/homebrew/bin/pdftotext'

# Path to the PDF directory
pdf_directory = '/Users/gr2757/Dropbox/global_taxation/Code/trying_git/tax_parsing/Outputs/'

# Loop through all countries
for country_folder in os.listdir(pdf_directory):
    country_folder_path = os.path.join(pdf_directory, country_folder)
    
    # Check if the country_folder is a directory
    if os.path.isdir(country_folder_path):
        # Create a new directory for the text files within each country's folder
        output_directory = os.path.join(country_folder_path, 'text_files')
        os.makedirs(output_directory, exist_ok=True)
        
        # Loop through all PDF files in the country's folder
        for pdf_file in os.listdir(country_folder_path):
            if pdf_file.endswith('.pdf'):
                pdf_path = os.path.join(country_folder_path, pdf_file)
                output_path = os.path.join(output_directory, pdf_file.replace('.pdf', '.txt'))
                
                # Command to run pdftotext
                result = subprocess.run([pdftotext_path, pdf_path, '-layout', output_path], capture_output=True, text=True)
                
                # Check the output
                if result.returncode == 0:
                    print(f'Converted {pdf_path} to text')
                else:
                    print(f'Error converting {pdf_path} to text:')
                    print(result.stderr)
```
